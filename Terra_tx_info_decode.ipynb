### Smart Contract request

```python
query_msg = 'link_available_for_payment'
contract = 'terra10kc4n52rk4xqny3hdew3ggjfk9r420pqxs9ylf'

terra.wasm.contract_query(contract,query_msg )
```

### Alternative

```python
import requests
link = 'terra1su6g4t4vwx7y0uh3ksancyaurj4l6w9pfs40qt'
query_msg = '{"balance":{"address":"' + contract + '"}}' #'{"decimals":{}}'
contract = 'terra10kc4n52rk4xqny3hdew3ggjfk9r420pqxs9ylf'

response = requests.get(
    LCDEndPoint+"/wasm/contracts/" + link + "/store",
    params={"query_msg": query_msg},
).json()

response
```

### Tx logs

```python
#latest block
block = terra.tendermint.block_info('latest')
#iterate over txs
for tx in block.get("block").get("data").get("txs"):
    txhash = hash_amino(tx)
    #tx info by hash
    print(terra.tx.tx_info(txhash)) 
    print('*******************************')  
```

```python
txinfo = terra.tx.tx_info(txhash)
txinfo.rawlog
```
