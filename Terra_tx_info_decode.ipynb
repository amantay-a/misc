```python
from terra_sdk.client.lcd import LCDClient
from terra_sdk.util.hash import hash_amino
terra = LCDClient(chain_id="columbus-5", url="https://lcd.terra.dev")

#latest block
block = terra.tendermint.block_info('latest')
#iterate over txs
for tx in block.get("block").get("data").get("txs"):
    txhash = hash_amino(tx)
    #tx info by hash
    print(terra.tx.tx_info(txhash)) 
    print('*******************************')  
```
