```python
import json
import requests
import pandas as pd
```

```python
url = 'https://api.yearn.finance/v1/chains/1/vaults/all'
params = {}
response = requests.get(url, params=params)
response_json = json.loads(response.text)
df = pd.DataFrame(response_json)
df
```

```python
df['gross_apr'] = df['apy'].apply(lambda x: x['gross_apr'])
df['net_apy'] = df['apy'].apply(lambda x: x['net_apy'])
df[['symbol', 'address', 'gross_apr', 'net_apy']]
```
